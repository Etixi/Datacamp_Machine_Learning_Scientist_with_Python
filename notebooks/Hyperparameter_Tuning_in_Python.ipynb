```python
credit_card = pd.read_csv('credit-card-full.csv')
# To change categorical variable with dummy variables
credit_card = pd.get_dummies(credit_card, columns=['SEX', 'EDUCATION', 'MARRIAGE'], drop_first=True)
credit_card.head()
```

```python
from sklearn.model_selection import train_test_split

X = credit_card.drop(['ID', 'default payment next month'], axis=1)
y = credit_card['default payment next month']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True)
```

```python
from sklearn.linear_model import LogisticRegression

log_reg_clf = LogisticRegression(max_iter=1000)
log_reg_clf.fit(X_train, y_train)

# Create a list of original variable names from the training DataFrame
original_variables = list(X_train.columns)

# Extract the coefficients of the logistic regression estimator
model_coefficients = log_reg_clf.coef_[0]

# Create a dataframe of the variables and coefficients & print it out
coefficient_df = pd.DataFrame({"Variable" : original_variables, "Coefficient": model_coefficients})
print(coefficient_df)
```

```python
# Print out the top 3 positive variables
top_three_df = coefficient_df.sort_values(by=["Coefficient"], axis=0, ascending=False)[0:3]
print(top_three_df)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import export_graphviz
import os
import pydot

rf_clf = RandomForestClassifier(max_depth=4, criterion='gini', n_estimators=10);
rf_clf.fit(X_train, y_train)

# Extract the 7th (index 6) tree from the random forest
chosen_tree = rf_clf.estimators_[6]

# Export with dot
export_graphviz(chosen_tree,
                out_file='tree6.dot',
                feature_names=X_train.columns,
                filled=True,
                rounded=True)
(graph, ) = pydot.graph_from_dot_file('tree6.dot')

# Convert dot to png
graph.write_png('tree_viz_image.png')

# Visualize the graph using the provided image
tree_viz_image = plt.imread('tree_viz_image.png')
plt.figure(figsize = (16,10))
plt.imshow(tree_viz_image, aspect='auto');
plt.axis('off')
```

```python
# Extract the parameters and level of the top (index 0) node
split_column = chosen_tree.tree_.feature[0]
split_column_name = X_train.columns[split_column]
split_value = chosen_tree.tree_.threshold[0]

# Print out the feature and level
print("This node split on feature {}, at a value of {}".format(split_column_name, split_value))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score

# Create the function
def gbm_grid_search(learn_rate, max_depth):
    # Create the model
    model = GradientBoostingClassifier(learning_rate=learn_rate, max_depth=max_depth)

    # Use the model to make predictions
    predictions = model.fit(X_train, y_train).predict(X_test)

    # Return the hyperparameters and score
    return ([learn_rate, max_depth, accuracy_score(y_test, predictions)])
```

####**5. Une astuce de probabilité**

+ ***Considérons la probabilité que nous continuions à manquer complètement les bons modèles, si nous sélectionnons au hasard des combinaisons d'hyperparamètres de manière uniforme Lors de notre premier essai, nous avons 5% de chances d'obtenir l'un de ces carrés car il s'agit de 5 carrés sur 100. Par conséquent, nous avons (1 moins 0-point-05) chance de rater ces cases. ####**6. Une astuce de probabilité**

####**1. Méthodes informées : statistiques bayésiennes**

####**2. Présentation de Bayes**

+ ***La règle de Bayes est un outil statistique célèbre qui existe depuis 250 ans. Alors, comment cela peut-il être pertinent pour l'apprentissage automatique aujourd'hui ? Eh bien, la règle de Bayes est en fait une méthode où nous pouvons utiliser de manière itérative de nouvelles preuves pour mettre à jour nos croyances sur un résultat. Intuitivement, cela a du sens car lors de l'utilisation de la recherche informée, nous voulons apprendre des preuves pour améliorer notre réglage des hyperparamètres.***

####**3. Règle de Bayes**

+ ***Voici la formule de la règle de Bayes. Le côté gauche est la probabilité de $A$ (qui nous tient à cœur), étant donné que $B$ s'est produit, où B est une nouvelle preuve (pertinente). C'est ce qu'on appelle le «postérieur». Le côté droit est la façon dont nous calculons cela. $P(A)$ est appelé "prior". C'est l'hypothèse initiale sur l'événement qui nous intéresse. Voyez en quoi c'est différent de $P(A|B)$ ? Cette dernière est la probabilité donnée de nouvelles preuves.***

####**4. Règle de Bayes**

+ ***$P(B)$ est la « vraisemblance marginale » et c'est la probabilité d'observer cette nouvelle preuve. $P(B|A)$ est la « vraisemblance », c'est-à-dire la probabilité d'observer la preuve, compte tenu de l'événement qui nous intéresse. Tout cela peut être assez déroutant, mais utilisons un exemple courant de diagnostic médical pour le démontrer.***

####**5. Bayes en médecine**

+ ***Prenons un exemple médical pour illustrer le processus bayésien. Disons que 5% des personnes dans la population générale ont une certaine maladie. C'est notre P(D). 10% des personnes sont génétiquement prédisposées à cette condition. Autrement dit, en raison de leur génétique, ils sont plus susceptibles de contracter cette maladie. C'est notre P(Pré). Nous savons également que 20 % des personnes atteintes de la maladie sont prédisposées, ce qui est notre P(Pré|D).***

####**6. Bayes en médecine**

+ ***Alors, quelle est la probabilité qu'une personne donnée soit atteinte de la maladie ? Si nous ne savons rien d'une personne, alors la probabilité qu'elle ait la maladie est juste la première. Cependant, que se passe-t-il si nous ajoutons de nouvelles preuves que cette personne est prédisposée ? Nous pouvons mettre à jour nos croyances en souscrivant à la formule de Bayes. Nous pouvons maintenant voir que la probabilité d'avoir la maladie, étant donné qu'une personne est prédisposée, est d'environ 10 %, soit le double de la probabilité initiale.***

####**7. Bayes dans le réglage des hyperparamètres**

```python
# Set up space dictionary with specified hyperparameters
space = {'max_depth': hp.hp.quniform('max_depth', 2, 10, 2),
         'learning_rate': hp.hp.uniform('learning_rate', 0.001, 0.9)}

# Set up objective function
def objective(params):
    params = {'max_depth': int(params['max_depth']),
              'learning_rate': params['learning_rate']}
    gbm_clf = GradientBoostingClassifier(n_estimators=100, **params)
    best_score = cross_val_score(gbm_clf, X_train, y_train,
                                 scoring='accuracy', cv=2, n_jobs=4).mean()
    loss = 1 - best_score
    return loss
```

```python
# Run the algorithm
best = hp.fmin(fn=objective,
            space=space,
            max_evals=20,
            rstate=np.random.default_rng(42),
            algo=hp.tpe.suggest)

print(best)
```

Des mutations aléatoires se produisent avec certains des descendants. Ces mutations contribuent parfois à donner un avantage à certains descendants. Enfin, retournez à (1) ! C'est ainsi que l'évolution fonctionne dans la nature.***

####**3. Génétique dans l'apprentissage automatique**

+ ***Nous pouvons appliquer la même idée au réglage des hyperparamètres. Tout d'abord, nous pouvons créer des modèles (qui ont des paramètres d'hyperparamètres). Deuxièmement, nous pouvons choisir le meilleur (grâce à notre fonction de notation). Ce sont ceux qui « survivent ». Nous pouvons alors créer de nouveaux modèles similaires aux meilleurs. Nous ajoutons un peu d'aléatoire afin de ne pas atteindre un optimum local. Enfin, nous continuons jusqu'à ce que nous soyons heureux!***

####**6. Composants TPOT**

+ ***Les arguments clés d'un classificateur TPOT sont : générations : Le nombre de cycles que nous entreprenons pour créer des modèles de progéniture, muter et traverser, choisir le meilleur et continuer. population_size : dans chaque itération, c'est le nombre de modèles que nous conservons. La "progéniture" la plus forte. offspring_size : à chaque itération, il s'agit du nombre de descendants (pour nous, de modèles) que nous créons. mutation_rate : Nous appliquons le caractère aléatoire à une proportion des pipelines. Cet hyperparamètre définit cette proportion (entre 0 et 1). crossover_rate : à chaque itération, nous croisons ou croisons certains de nos modèles pour en trouver des similaires. Cela définit la proportion de pipelines auxquels nous faisons cela. notation : la fonction objectif pour déterminer les modèles ou les descendants les plus forts, par exemple, la précision.
