```python
# Set the learning rate: learning_rate
learning_rate = 0.01

# Calculate the predictions: preds
preds = (weights * input_data).sum()

# Calculate the error: error
error = preds - target

# Calculate the slope: slope
slope = 2 * input_data * error

# Update the weights: weights_updated
weights_updated = weights - learning_rate * slope

# Get updated predictions: preds_updated
preds_updated = (input_data*weights_updated).sum()

# Calculate updated error: error_updated
error_updated = preds_updated - target

# Print the original error
print(error)

# Print the updated error
print(error_updated)
```

####**Effectuer plusieurs mises à jour des poids**

+ ***Vous allez maintenant effectuer plusieurs mises à jour afin d'améliorer considérablement les pondérations de votre modèle et de voir comment les prédictions s'améliorent à chaque mise à jour.***

+ ***Pour garder votre code propre, il existe une fonction préchargée `get_slope()` qui prend input_data, target et weights comme arguments. Il existe également une fonction `get_mse()` qui prend les mêmes arguments. Les `input_data, target et weights` ont été préchargés.***

+ ***Ce réseau n'a pas de couches cachées, et il va directement de l'entrée (avec 3 nœuds) à un nœud de sortie. Notez qu'il s'agit d'un seul tableau weights.**

+ ***Nous avons également pré-chargé matplotlib.pyplot et l'historique des erreurs sera tracé une fois que vous aurez effectué vos étapes de descente de gradient.***

####**Instructions**

+ Utilisation d'une boucle for pour mettre à jour les pondérations de manière itérative :
  + Calculez la pente à l'aide de la fonction get_slope().
  + Mettez à jour les pondérations en utilisant un taux d'apprentissage de 0.01.
  + Calculez l'erreur quadratique moyenne (mse) avec les poids mis à jour à l'aide de la fonction get_mse().
  + Ajouter mse à mse_hist.
  + Cliquez sur "Soumettre la réponse" pour visualiser mse_hist. Quelle tendance remarquez-vous ?

```python
def get_error(input_data, target, weights):
    preds = (weights * input_data).sum()
    error = preds - target
    return error

def get_slope(input_data, target, weights):
    error = get_error(input_data, target, weights)
    slope = 2 * input_data * error
    return slope

def get_mse(input_data, target, weights):
    errors = get_error(input_data, target, weights)
    mse = np.mean(errors ** 2)
    return mse
```

```python
n_updates = 20
mse_hist = []

# Iterate over the number of updates
for i in range(n_updates):
    # Calculate the slope: slope
    slope = get_slope(input_data, target, weights)

    # Update the weights: weights
    weights = weights- slope * 0.01

    # Calculate mse with new weights: mse
    mse = get_mse(input_data, target, weights)

    # Append the mse to mse_hist
    mse_hist.append(mse)

# Plot the mse history
plt.plot(mse_hist)
plt.xlabel('Iterations')
plt.ylabel('Mean Squared Error')
plt.show()
```

+ ***problème de "neurone mourant". Ce problème survient lorsqu'un neurone prend une valeur inférieure à 0 pour toutes les lignes de vos données. En d'autres termes, une fois que le nœud commence à toujours recevoir des entrées négatives, il peut continuer à ne recevoir que des entrées négatives. Lorsque vous surajustez, votre modèle fera des prédictions précises sur les données de formation, mais il fera des prédictions inexactes sur les données de validation et les nouveaux ensembles de données. Le sous-ajustement est le contraire. C'est alors que votre modèle ne parvient pas à trouver des modèles prédictifs importants dans les données d'apprentissage. Il n'est donc précis ni dans les données de formation ni dans les données de validation.
