####**Instructions**


+ Définissez features, params, et bill comme constantes.
+ Calculez le vecteur de valeur prédite, billpred, en multipliant les données d'entrée, features, par les paramètres, params.
+ Utilisez la multiplication matricielle plutôt que le produit élément par élément.

####**6. Calcul du gradient**

+ ***Ce n'est pas vrai si nous considérons plutôt la fonction $y$ égale $x$ au carré. Lorsque $x$ est inférieur à 0, $y$ diminue lorsque $x$ augmente. Cela signifie que $x$ égal à 0 minimise $y$.***

####**Optimiser avec des dégradés**

+ ***On vous donne une fonction de perte,$y = x^{2}$
, que vous souhaitez minimiser. Vous pouvez le faire en calculant la pente en utilisant l' opération `GradientTape()` à différentes valeurs de $x$. Si la pente est positive, vous pouvez diminuer la perte en diminuant $x$. S'il est négatif, vous pouvez le diminuer en augmentant $x$. C'est ainsi que fonctionne la descente de gradient.***

+ ***L'image montre un tracé de $y$ égal $x$ au carré.  Il montre également le gradient à $x$ égal à -1, $x$ égal à 0 et $x$ égal à 1.***

<img src="https://assets.datacamp.com/production/repositories/3953/datasets/4a3d06616c28aed697d57914a26da3d831bac83c/gradient_plot.png"/>

+ ***En pratique, vous utiliserez une opération `tensorflow` de haut niveau pour effectuer automatiquement une descente de gradient. Dans cet exercice, cependant, vous calculerez la pente aux xvaleurs -1, 1 et 0. Les opérations suivantes sont disponibles : `GradientTape(), multiply() et Variable()`.***

####**Instructions**

+ Définir $x$ comme une variable avec la valeur initiale $x0$.
+ Définissez la fonction de perte, $y$, égale à $x$ multipliée par $x$. N'utilisez pas la surcharge de l'opérateur.
+ Définissez la fonction pour renvoyer le gradient de $y$ par rapport à $x$.

```python
def compute_gradient(x0):
  	# Define x as a variable with an initial value of x0
	x = tf.Variable(x0)
	with tf.GradientTape() as tape:
		tape.watch(x)
        # Define y using the multiply operation
		y = tf.multiply(x,x)
    # Return the gradient of y with respect to x
	return tape.gradient(y, x).numpy()
```

```python
# Compute and print gradients at x = -1, 1, and 0
print(compute_gradient(-1.0))
print(compute_gradient(1.0))
print(compute_gradient(0.0))
```

***Excellent travail! Notez que la pente est positive à $x= 1$, ce qui signifie que nous pouvons réduire la perte en réduisant $x$. La pente est négative à $x= -1$, ce qui signifie que nous pouvons réduire la perte en augmentant $x$. La pente à $x= 0$ est 0, ce qui signifie que nous ne pouvons pas réduire la perte en augmentant ou en diminuant $x$. C'est parce que la perte est minimisée à $x= 0$.***

```python
# Define the model
def model(w1, b1, w2, b2, features=borrower_features):
    # Apply relu activation function to layer 1
    layer1 = tf.keras.activations.relu(tf.matmul(features, w1) + b1)

    # Apply Dropout
    dropout = tf.keras.layers.Dropout(0.25)(layer1)
    return tf.keras.activations.sigmoid(tf.matmul(dropout, w2) + b2)

# Define the loss function
def loss_function(w1, b1, w2, b2, features=borrower_features, targets = borrower_targets):
    predictions = model(w1, b1, w2, b2)

    # Pass targets and predictions to the cross entropy loss
    return tf.keras.losses.binary_crossentropy(targets, predictions)
```
