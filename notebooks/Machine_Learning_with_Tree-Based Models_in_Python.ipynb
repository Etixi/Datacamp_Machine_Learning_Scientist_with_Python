```python
from mlxtend.plotting import plot_decision_regions

def plot_labeled_decision_regions(X,y, models):
    '''Function producing a scatter plot of the instances contained
    in the 2D dataset (X,y) along with the decision
    regions of two trained classification models contained in the
    list 'models'.

    Parameters
    ----------
    X: pandas DataFrame corresponding to two numerical features
    y: pandas Series corresponding the class labels
    models: list containing two trained classifiers

    '''
    if len(models) != 2:
        raise Exception('''Models should be a list containing only two trained classifiers.''')
    if not isinstance(X, pd.DataFrame):
        raise Exception('''X has to be a pandas DataFrame with two numerical features.''')
    if not isinstance(y, pd.Series):
        raise Exception('''y has to be a pandas Series corresponding to the labels.''')
    fig, ax = plt.subplots(1, 2, figsize=(10.0, 5), sharey=True)
    for i, model in enumerate(models):
        plot_decision_regions(X.values, y.values, model, legend= 2, ax = ax[i])
        ax[i].set_title(model.__class__.__name__)
        ax[i].set_xlabel(X.columns[0])
        if i == 0:
            ax[i].set_ylabel(X.columns[1])
            ax[i].set_ylim(X.values[:,1].min(), X.values[:,1].max())
            ax[i].set_xlim(X.values[:,0].min(), X.values[:,0].max())
    plt.tight_layout()
```

####**2. Éléments constitutifs d'un arbre décisionnel**

+ ***Commençons d'abord par définir quelques termes. Un arbre de décision est une structure de données composée d'une hiérarchie d'unités individuelles appelées nœuds. Un nœud est un point qui implique soit une question, soit une prédiction.***

####**3. Éléments constitutifs d'un arbre décisionnel**

+ ***La racine est le nœud auquel l'arbre de décision commence à croître. Il n'a pas de nœud parent et implique une question qui donne lieu à 2 nœuds enfants à travers deux branches. Un nœud interne est un nœud qui a un parent. Il s'agit également d'une question qui donne lieu à 2 nœuds enfants. Enfin, un nœud qui n'a pas d'enfant est appelé une feuille. Une feuille a un nœud parent et n'implique aucune question. C'est là qu'une prédiction est faite. Rappelez-vous que lorsqu'un arbre de classification est formé sur un jeu de données étiqueté, l'arbre apprend des modèles à partir des caractéristiques de manière à produire les feuilles les plus pures. En d'autres termes, l'arbre est formé de manière à ce que, dans chaque feuille, une étiquette de classe soit prédominante.***

####**4. Prédiction**

+ ***Dans le diagramme en arbre présenté ici, considérons le cas où une instance traverse l'arbre pour atteindre la feuille de gauche. Dans cette feuille, il y a 257 cas classés comme bénins et 7 cas classés comme malins. Par conséquent, la prédiction de l'arbre pour cette instance serait : « bénigne ». Afin de comprendre comment un arbre de classification produit les feuilles les plus pures possibles, définissons d'abord le concept de gain d'information.***

####**5. Gain d'informations (IG)**

+ ***Les nœuds d'un arbre de classification sont développés de manière récursive ; autrement dit, l'obtention d'un nœud interne ou d'une feuille dépend de l'état de ses prédécesseurs. Pour produire les feuilles les plus pures possibles, à chaque nœud, un arbre pose une question impliquant une caractéristique f et un point de division sp. Mais comment sait-il quelle fonctionnalité et quel point de partage choisir ? Il le fait en maximisant le gain d'informations ! L'arbre considère que chaque nœud contient de l'information et vise à maximiser le gain d'information obtenu après chaque division. Considérons le cas où un nœud avec N échantillons est divisé en un nœud gauche avec N échantillons gauches et un nœud droit avec N échantillons droits.***

####**6. Gain d'informations (IG)**

+ ***Le gain d'information pour une telle division est donné par la formule présentée ici. Une question que vous pourriez avoir à l'esprit ici est : 'Quel critère est utilisé pour mesurer l'impureté d'un nœud ?' Eh bien, il existe différents critères que vous pouvez utiliser, parmi lesquels l'indice de Gini et l'entropie. Maintenant que vous savez ce qu'est le gain d'information, décrivons comment un arbre de classification apprend.***

####**7. Apprentissage de l'arbre de classification**

+ ***Lorsqu'un arbre sans contrainte est formé, les nœuds sont développés de manière récursive. En d'autres termes, un nœud existe en fonction de l'état de ses prédécesseurs. Au niveau d'un nœud non-feuille, les données sont divisées en fonction de la caractéristique f et du point de division sp de manière à maximiser le gain d'informations. Si le gain d'information obtenu en divisant un nœud est nul, le nœud est déclaré feuille. Gardez à l'esprit que ces règles s'appliquent aux arbres non contraints. Si vous contraignez la profondeur maximale d'un arbre à 2 par exemple, tous les nœuds ayant une profondeur de 2 seront déclarés feuilles même si le gain d'information obtenu en divisant ces nœuds n'est pas nul.***

####**2. Apprentissage supervisé - Sous le capot**

+ ***Dans l'apprentissage supervisé, vous faites l'hypothèse qu'il existe une correspondance $f$ entre les fonctionnalités et les étiquettes. Vous pouvez l'exprimer sous la forme $y=f(x)$. $f$ qui est affiché en rouge ici est une fonction inconnue que vous voulez déterminer. En réalité, la génération de données s'accompagne toujours d'un caractère aléatoire ou de bruit comme les points bleus illustrés ici.***

####**3. Objectifs de l'apprentissage supervisé**

+ ***Votre but est de trouver un modèle $\hat{f}$ qui se rapproche le mieux de $f$. Lors de l'entraînement, vous voulez vous assurer que le bruit est éliminé autant que possible. À la fin, cela devrait permettre d'obtenir une faible erreur prédictive sur des ensembles de données invisibles.***

####**4. Difficultés à approximer $f$**

+ ***Vous pouvez rencontrer deux difficultés lors de l'approximation de $f$. `Le premier est le surajustement, c'est quand ça correspond au bruit dans l'ensemble d'entraînement. Le second est le sous-ajustement, c'est quand $f$ n'est pas assez flexible pour approximer $f$.`***

####**5. Sur-ajustement**

+ ***Lorsqu'un modèle dépasse l'ensemble d'apprentissage, son pouvoir prédictif sur des ensembles de données invisibles est assez faible. Ceci est illustré par les prédictions du régresseur de l'arbre de décision représenté ici en rouge. Le modèle a clairement mémorisé le bruit présent dans l'ensemble d'apprentissage. Un tel modèle atteint une faible erreur d'ensemble d'apprentissage et une erreur d'ensemble de test élevée.***

####**6. Sous-équipement**

+ ***Lorsqu'un modèle est sous-ajusté aux données, comme l'arbre de régression dont les prédictions sont affichées ici en rouge, l'erreur de l'ensemble d'apprentissage est à peu près égale à l'erreur de l'ensemble de test. Cependant, les deux erreurs sont relativement élevées. Désormais, le modèle formé n'est pas suffisamment flexible pour capturer la dépendance complexe entre les fonctionnalités et les étiquettes. Par analogie, c'est comme enseigner le calcul à un enfant de 3 ans. L'enfant n'a pas le niveau d'abstraction mentale requis pour comprendre le calcul.***

###**7. Erreur de généralisation**

+ ***L'erreur de généralisation d'un modèle vous indique à quel point il généralise sur des données invisibles. Il peut être décomposé en 3 termes : `biais, variance et erreur irréductible où l'erreur irréductible est la contribution du bruit à l'erreur.`***

####**8. Biais**

+ ***Le terme de biais vous indique, en moyenne, à quel point $\hat{f}$ et $f$ sont différents. Pour illustrer cela, considérons le modèle à fort biais représenté ici en noir ; ce modèle n'est pas assez flexible pour se rapprocher de la vraie fonction $f$ indiquée en rouge. Les modèles à fort biais conduisent à un sous-ajustement.***

####**9. Écart**

+ ***Le terme de variance vous indique dans quelle mesure ce qui est incohérent sur différents ensembles d'apprentissage. Considérez le modèle à forte variance représenté ici en noir ; dans ce cas, fhat suit si étroitement les points de données d'apprentissage qu'il manque la vraie fonction $f$ indiquée en rouge. Les modèles à forte variance conduisent à un surajustement.***

####**10. Complexité du modèle**

+ ***La complexité d'un modèle détermine sa flexibilité pour approximer la vraie fonction $f$. Par exemple : l'augmentation de la profondeur maximale de l'arbre augmente la complexité d'un arbre de décision.***

####**11. Compromis biais-variance**

+ ***Le diagramme ici montre comment la meilleure complexité du modèle correspond à l'erreur de généralisation la plus faible. Lorsque la complexité du modèle augmente, la variance augmente tandis que le biais diminue. Inversement, lorsque la complexité du modèle diminue, la variance diminue et le biais augmente. Votre objectif est de trouver la complexité du modèle qui atteint l'erreur de généralisation la plus faible. Étant donné que cette erreur est la somme de trois termes, l'erreur irréductible étant constante, vous devez trouver un équilibre entre le biais et la variance, car à mesure que l'un augmente, l'autre diminue. C'est ce qu'on appelle le compromis biais-variance.***

####**12. Compromis biais-variance : une explication visuelle**

+ ***Visuellement, vous pouvez imaginer approximer fhat comme visant le centre d'une cible de tir où le centre est la vraie fonction f. S'il s'agit d'un faible biais et d'une faible variance, vos prises de vue seront étroitement regroupées autour du centre. S'il s'agit d'une variance et d'un biais élevés, non seulement vos tirs manqueront la cible, mais ils seront également répartis tout autour de la cible de tir.***

De plus, les CART sont faciles à utiliser et leur flexibilité leur permet de décrire les dépendances non linéaires entre les fonctionnalités et les étiquettes. De plus, vous n'avez pas besoin de beaucoup de prétraitement de fonctionnalités pour former un CART. Contrairement à d'autres modèles, vous n'avez pas besoin de standardiser ou de normaliser les fonctionnalités avant de les alimenter dans un CART.***

####**3. Limites des CART**

+ ***Les chariots ont également des limites. Un arbre de classification, par exemple, ne peut produire que des frontières de décision orthogonales. Les CART sont également très sensibles aux petites variations de l'ensemble d'apprentissage. Parfois, lorsqu'un seul point est supprimé de l'ensemble d'apprentissage, les paramètres appris d'un CART peuvent changer radicalement. Les CART souffrent également d'une variance élevée lorsqu'ils sont entraînés sans contraintes. Dans ce cas, ils peuvent sur-adapter l'ensemble d'entraînement. Une solution qui tire parti de la flexibilité des CART tout en réduisant leur tendance à mémoriser le bruit est l'apprentissage d'ensemble.***

####**4. Apprentissage d'ensemble**

+ ***L'apprentissage d'ensemble peut être résumé comme suit : - Dans un premier temps, différents modèles sont entraînés sur le même jeu de données. -Chaque modèle fait ses propres prédictions. -Un méta-modèle agrège ensuite les prédictions des modèles individuels et produit une prédiction finale. -La prédiction finale est plus robuste et moins sujette aux erreurs que chaque modèle individuel. -Les meilleurs résultats sont obtenus lorsque les modèles sont habiles mais de manière différente ce qui signifie que si certains modèles font des prédictions très éloignées, les autres modèles devraient compenser ces erreurs. Dans ce cas, les prédictions du méta-modèle sont plus robustes.***

####**5. Apprentissage d'ensemble : une explication visuelle**

+ ***Jetons un coup d'œil au diagramme ici pour comprendre visuellement comment fonctionne l'apprentissage d'ensemble pour un problème de classification. Tout d'abord, l'ensemble d'apprentissage est envoyé à différents classificateurs. Chaque classificateur apprend ses paramètres et fait des prédictions. Ensuite, ces prédictions sont transmises à un méta-modèle qui les agrège et produit une prédiction finale.***

####**1. Ada Boost**

+ ***Le boosting fait référence à une méthode d'ensemble dans laquelle de nombreux prédicteurs sont formés et chaque prédicteur apprend des erreurs de son prédécesseur.***

####**2. Booster**

+ ***Plus formellement, en stimulant de nombreux apprenants faibles sont combinés pour former un apprenant fort. Un apprenant faible est un modèle qui fait un peu mieux que la supposition aléatoire. Par exemple, un arbre de décision avec une profondeur maximale de un, appelé souche de décision, est un apprenant faible.***

####**3. Booster**

+ ***Dans le boosting, un ensemble de prédicteurs est formé séquentiellement et chaque prédicteur essaie de corriger les erreurs commises par son prédécesseur. Les deux méthodes d'amplification que vous explorerez dans ce cours sont AdaBoost et Gradient Boosting.***

####**4. Adaboost**

+ ***AdaBoost signifie Adaptive Boosting. Dans AdaBoost, chaque prédicteur accorde plus d'attention aux instances mal prédites par son prédécesseur en changeant constamment les poids des instances d'entraînement. De plus, chaque prédicteur se voit attribuer un coefficient alpha qui pondère sa contribution dans la prédiction finale de l'ensemble. Alpha dépend de l'erreur d'apprentissage du prédicteur.***

####**5. AdaBoost : Entraînement**

+ ***Comme le montre le diagramme, il y a $N$ prédicteurs au total. Tout d'abord, le prédicteur1 est formé sur l'ensemble de données initial $(X,y)$ et l'erreur de formation pour le prédicteur1 est déterminée. Cette erreur peut ensuite être utilisée pour déterminer alpha1 qui est le coefficient du prédicteur1. Alpha1 est ensuite utilisé pour déterminer les poids $W(2)$ des instances d'entraînement pour le prédicteur2. Remarquez comment les instances mal prédites affichées en vert acquièrent des poids plus élevés. Lorsque les instances pondérées sont utilisées pour former le prédicteur2, ce prédicteur est obligé de prêter plus d'attention aux instances mal prédites. Ce processus est répété séquentiellement, jusqu'à ce que les $N$ prédicteurs formant l'ensemble soient entraînés.***

####**6. Taux d'apprentissage**
+ ***Un paramètre important utilisé dans la formation est le taux d'apprentissage, eta. Eta est un nombre compris entre 0 et 1 ; il est utilisé pour réduire le coefficient alpha d'un prédicteur formé. Il est important de noter qu'il existe un compromis entre eta et le nombre d'estimateurs. Une plus petite valeur d'êta devrait être compensée par un plus grand nombre d'estimateurs.***

####**7. AdaBoost : Prédiction**

+ ***Une fois que tous les prédicteurs de l'ensemble sont formés, l'étiquette d'une nouvelle instance peut être prédite en fonction de la nature du problème. Pour la classification, chaque prédicteur prédit l'étiquette de la nouvelle instance et la prédiction de l'ensemble est obtenue par vote à la majorité pondérée. Pour la régression, la même procédure est appliquée et la prédiction de l'ensemble est obtenue en effectuant une moyenne pondérée. Il est important de noter que les prédicteurs individuels n'ont pas besoin d'être des CART. Cependant, les CART sont utilisés la plupart du temps en boosting en raison de leur forte variance.***

####**3. Boost de gradient stochastique**

+ ***Pour atténuer ces effets, vous pouvez utiliser un algorithme connu sous le nom de renforcement de gradient stochastique. Dans le renforcement du gradient stochastique, chaque CART est formé sur un sous-ensemble aléatoire des données de formation. Ce sous-ensemble est échantillonné sans remise. De plus, au niveau de chaque nœud, les caractéristiques sont échantillonnées sans remise lors du choix des meilleurs points de partage. En conséquence, cela crée une plus grande diversité dans l'ensemble et l'effet net ajoute plus de variance à l'ensemble d'arbres.***

####**4. Amplification du gradient stochastique : entraînement**

+ ***Examinons de plus près la procédure d'entraînement utilisée dans l'amplification du gradient stochastique en examinant le diagramme présenté sur cette diapositive. Premièrement, au lieu de fournir toutes les instances d'apprentissage à un arbre, seule une fraction de ces instances est fournie par échantillonnage sans remplacement. Les données échantillonnées sont ensuite utilisées pour former un arbre. Cependant, toutes les fonctionnalités ne sont pas prises en compte lors d'une scission. Au lieu de cela, seule une certaine fraction échantillonnée au hasard de ces caractéristiques est utilisée à cette fin. Une fois qu'un arbre est formé, des prédictions sont faites et les erreurs résiduelles peuvent être calculées. Ces erreurs résiduelles sont multipliées par le taux d'apprentissage eta et transmises à l'arbre suivant de l'ensemble. Cette procédure est répétée séquentiellement jusqu'à ce que tous les arbres de l'ensemble soient entraînés.***

####**2. Hyperparamètres**

+ ***Les modèles d'apprentissage automatique sont caractérisés par des paramètres et des hyperparamètres. Les paramètres sont appris à partir des données grâce à la formation ; des exemples de paramètres incluent la fonctionnalité de fractionnement et le point de partage d'un nœud dans un CART. Les hyperparamètres ne sont pas appris à partir des données ; ils doivent être définis avant la formation. Des exemples d'hyperparamètres incluent la profondeur maximale et le critère de fractionnement d'un CART.***

####**3. Qu'est-ce que le réglage des hyperparamètres ?**

+ ***Le réglage des hyperparamètres consiste à rechercher l'ensemble des hyperparamètres optimaux pour l'algorithme d'apprentissage. La solution consiste à trouver l'ensemble d'hyperparamètres optimaux produisant un modèle optimal. Le modèle optimal donne un score optimal. La fonction de score mesure l'accord entre les véritables étiquettes et les prédictions d'un modèle. Dans sklearn, la valeur par défaut est la précision pour les classificateurs et le r au carré pour les régresseurs. Les performances de généralisation d'un modèle sont évaluées à l'aide d'une validation croisée.***

####**4. Pourquoi régler les hyperparamètres ?**

+ ***Une question légitime que vous pouvez vous poser est : pourquoi s'embêter à régler les hyperparamètres ? Eh bien, dans scikit-learn, les hyperparamètres par défaut d'un modèle ne sont pas optimaux pour tous les problèmes. Les hyperparamètres doivent être réglés pour obtenir les meilleures performances du modèle.***

####**5. Approches du réglage des hyperparamètres**

+ ***Il existe maintenant de nombreuses approches pour le réglage des hyperparamètres, notamment : la recherche sur grille, la recherche aléatoire, etc. Dans ce cours, nous explorerons uniquement la méthode de recherche par grille.***

####**6. Validation croisée de la recherche par grille**

+ ***Dans la validation croisée de recherche de grille, vous définissez d'abord manuellement une grille de valeurs d'hyperparamètres discrètes. Ensuite, vous choisissez une métrique pour noter les performances du modèle et vous effectuez une recherche exhaustive dans la grille. Pour chaque ensemble d'hyperparamètres, vous évaluez le score de chaque modèle. Les hyperparamètres optimaux sont ceux pour lesquels le modèle obtient le meilleur score de validation croisée. Notez que la recherche de grille souffre de la malédiction de la dimensionnalité, i-dot-e-dot, plus la grille est grande, plus il faut de temps pour trouver la solution.***

####**7. Validation croisée de la recherche par grille : exemple**

+ ***Passons en revue un exemple concret pour comprendre cette procédure.
