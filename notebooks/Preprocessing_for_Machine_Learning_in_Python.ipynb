####**1. Ensembles d'entraînement et de test**

+ ***L'une des étapes clés du prétraitement que vous devez connaître consiste à diviser les données en ensembles d'apprentissage et de test.***

####**2. Pourquoi diviser ?**

+ ***Nous avons divisé notre ensemble de données en formation et en test pour quelques raisons principales. Premièrement, cela réduit le risque de surajustement, qui, rappelons-le, survient lorsque le modèle s'adapte trop étroitement aux données d'apprentissage, ce qui entraîne de mauvaises performances lors de la prédiction sur des données invisibles. Deuxièmement, si nous formons un modèle sur l'ensemble de notre ensemble de données, nous n'aurons aucun moyen de tester et de valider notre modèle, car le modèle connaîtra essentiellement l'ensemble de données par cœur. La tenue d'un jeu de test nous permet de conserver certaines données que le modèle n'a pas encore vues, afin que nous puissions évaluer les performances du modèle sur des données invisibles.***

####**3. Fractionner votre jeu de données**

+ ***La fonction train_test_split de sklearn model_selection est utilisée pour mélanger de manière aléatoire, puis diviser les fonctionnalités et les étiquettes, stockées dans $X$ et $y$, en ensembles d'apprentissage et de test. X_train et X_test sont les fonctionnalités d'entraînement et de test, et y_train et y_test sont les étiquettes d'entraînement et de test. C'est une bonne pratique de spécifier l'argument random_state, afin que nous puissions reproduire exactement les mêmes fractionnements si nécessaire. Par défaut, la fonction divisera 75 % des données dans l'ensemble d'apprentissage et 25 % dans l'ensemble de test, mais nous pouvons ajuster la proportion des données affectées à l'ensemble de test avec l'argument test_size. Dans de nombreux scénarios, les paramètres de fractionnement par défaut fonctionneront bien. Cependant, si nos étiquettes ont une distribution inégale, où une étiquette est beaucoup plus courante qu'une autre, les ensembles de test et d'apprentissage peuvent ne pas être des échantillons représentatifs de l'ensemble de données, ce qui pourrait biaiser le modèle que nous essayons de former. C'est ce qu'on appelle le déséquilibre des classes. Par exemple, dans la formation et le test présentés ici, nous pouvons voir que l'ensemble d'apprentissage ne contient que des échantillons étiquetés n, alors qu'il y a une étiquette $y$ dans l'ensemble de test.***

####**4. Échantillonnage stratifié**

+ ***Une bonne technique pour échantillonner plus précisément lorsque vous avez des classes déséquilibrées est l'échantillonnage stratifié, qui est une méthode d'échantillonnage qui prend en compte la distribution des classes dans l'ensemble de données. Disons que nous avons un ensemble de données avec 100 échantillons, dont 80 sont de classe 1 et 20 sont de classe 2. Nous voulons que la distribution des classes dans notre ensemble d'entraînement et notre ensemble de test reflète cela, donc dans nos ensembles d'entraînement et de test , nous voudrions que 80 % de notre échantillon soit de classe 1 et 20 % de classe 2, ce qui signifie que nous voudrions 60 échantillons de classe 1 et 15 échantillons de classe 2 dans notre ensemble d'apprentissage de 75 échantillons. Dans notre ensemble de test de 25 échantillons, nous voulons avoir 20 échantillons de classe 1 et 5 de classe 2. Cela correspond à la distribution des classes dans l'ensemble de données d'origine.***

####**5. Échantillonnage stratifié**

+ ***Il existe une bonne façon de le faire en utilisant la fonction train_test_split. La fonction a un paramètre de stratification, et pour stratifier en fonction des étiquettes de classe, transmettez les étiquettes de l'ensemble de données, y, à cet argument. Le jeu de données contient 100 étiquettes, dont 80 sont de classe 1 et 20 sont de classe 2.***

####**1. Normalisation**

+ ***Il est possible que vous rencontriez des ensembles de données avec beaucoup de bruit numérique, peut-être en raison de la variance des caractéristiques ou de données mises à l'échelle différemment. La solution de prétraitement pour cela est la standardisation.***

####**2. Qu'est-ce que la normalisation ?**

+ ***La standardisation est une méthode de prétraitement utilisée pour transformer des données continues afin qu'elles paraissent normalement distribuées. Dans scikit-learn, il s'agit souvent d'une étape nécessaire, car de nombreux modèles supposent sous-jacents que les données d'apprentissage sont normalement distribuées, et si ce n'est pas le cas, nous risquons de biaiser votre modèle. Les données peuvent être standardisées de différentes manières, mais dans ce cours, nous allons parler de deux méthodes : la normalisation du journal et la mise à l'échelle. Il est également important de noter que la normalisation est une méthode de prétraitement appliquée à des données numériques continues. Nous aborderons les méthodes de traitement des données catégorielles plus tard dans le cours.***

####**3. Quand normaliser : distances linéaires**

+ ***Il existe plusieurs scénarios différents dans lesquels nous voudrions standardiser vos données. Tout d'abord, si nous travaillons avec n'importe quel type de modèle qui utilise une métrique de distance linéaire ou fonctionne dans un espace linéaire comme les k-plus proches voisins, la régression linéaire ou le clustering k-means, le modèle suppose que les données et les caractéristiques que nous ' re donnant sont liés de manière linéaire, ou peuvent être mesurés avec une métrique de distance linéaire, ce qui peut ne pas toujours être le cas.***

####**4. Quand standardiser : variance élevée**

+ ***La normalisation doit également être utilisée lorsque les entités du jeu de données ont une variance élevée, qui est également liée aux métriques de distance. Cela pourrait biaiser un modèle qui suppose que les données sont normalement distribuées. Si une entité de notre jeu de données présente une variance d'un ordre de grandeur ou plus que les autres caractéristiques, cela pourrait avoir un impact sur la capacité du modèle à apprendre des autres caractéristiques de l'ensemble de données.***

####**5. Quand standardiser : différentes échelles**

+ ***La modélisation d'un jeu de données contenant des entités continues à différentes échelles est un autre scénario de normalisation. Par exemple, envisagez de prédire les prix des maisons à l'aide de deux caractéristiques : le nombre de chambres et le dernier prix de vente. Ces deux caractéristiques sont à des échelles très différentes, ce qui confondra la plupart des modèles. Pour comparer ces caractéristiques, il faut les standardiser pour les mettre dans le même espace linéaire. Tous ces scénarios supposent que nous travaillons avec un modèle qui fait une sorte d'hypothèses de linéarité ; cependant, il existe un certain nombre de modèles qui fonctionnent parfaitement bien dans un espace non linéaire, ou qui effectuent une certaine normalisation lors de l'entrée, mais ils sortent du cadre de ce cours.***

```python
import re

# Write a pattern to extract numbers and decimals
def return_mileage(length):
    pattern = re.compile(r'\d+\.\d+')

    if length == None:
        return

    # Search the text for matches
    mile = re.match(pattern, length)

    # If a value is returned, use group(0) to return the found value
    if mile is not None:
        return float(mile.group(0))

# Apply the function to the Length column and take a look at both columns
hiking['Length_num'] = hiking['Length'].apply(lambda row: return_mileage(row))
hiking[['Length', 'Length_num']].head()
```

```python
# Add in the rest of the parameters
def return_weights(vocab, original_vocab, vector, vector_index, top_n):
    zipped = dict(zip(vector[vector_index].indices, vector[vector_index].data))

    # Let's transform that zipped dict into a series
    zipped_series = pd.Series({vocab[i]:zipped[i] for i in vector[vector_index].indices})

    # Let's sort the series to pull out the top n weighted words
    zipped_index = zipped_series.sort_values(ascending=False)[:top_n].index
    return [original_vocab[i] for i in zipped_index]
```

```python
# Print out the weighted words
print(return_weights(vocab, tfidf_vec.vocabulary_, text_tfidf, vector_index=8, top_n=3))
```

```python
def words_to_filter(vocab, original_vocab, vector, top_n):
    filter_list = []
    for i in range(0, vector.shape[0]):

        # Here we'll call the function from the previous exercise, and extend the list we're creating
        filtered = return_weights(vocab, original_vocab, vector, i, top_n)
        filter_list.extend(filtered)
    # Return the list in a set, so we don't get duplicate word indices
    return set(filter_list)
```

####**2. Réduction de la dimensionnalité et PCA**

+ ***La réduction de la dimensionnalité est une forme d'apprentissage non supervisé qui transforme nos données de manière à réduire le nombre de caractéristiques dans l'espace des caractéristiques. Cette transformation de données peut être effectuée de manière linéaire ou non linéaire. La réduction de la dimensionnalité est vraiment une méthode d'extraction de caractéristiques, puisque ces données sont transformées en caractéristiques nouvelles et différentes. Cependant, puisque nous le traitons ici comme une réduction de notre espace de fonctionnalités, nous le couvrirons dans ce chapitre. La méthode de réduction de dimensionnalité que nous aborderons est l'analyse en composantes principales, ou PCA. PCA utilise une transformation linéaire pour projeter des caractéristiques dans un espace où elles sont complètement non corrélées. Alors que l'espace des caractéristiques est réduit, la variance est capturée de manière significative en combinant les caractéristiques en composants. PCA capte, dans chaque composant, autant de variance dans l'ensemble de données que possible. En termes de sélection de fonctionnalités, cela peut être une méthode utile lorsque nous avons un grand nombre de fonctionnalités et aucun candidat solide pour l'élimination.***

####**3. PCA dans scikit-learn**

+ ***La transformation d'un ensemble de données via PCA est relativement simple dans scikit-learn. Semblable à d'autres méthodes d'apprentissage automatique, scikit-learn nécessite l'importation de pca et la création de l'objet pca. Et tout comme nous l'avons fait pour créer des vecteurs tf-idf, nous pouvons utiliser la méthode fit_transform de PCA sur l'ensemble de données sur lequel nous voulons réduire la dimensionnalité. Si nous imprimons le nouveau vecteur transformé PCA, nous pouvons voir que les données ont été transformées. Par défaut, PCA dans scikit-learn maintient le nombre de composants égal au nombre d'entités d'entrée. Si nous imprimons le ratio de variance expliquée, nous pouvons voir, par composante, le pourcentage de variance expliquée par cette composante. Nous pouvons voir qu'une grande partie de la variance est expliquée par le premier composant ici, il est donc probable que nous pourrions laisser tomber les composants qui n'expliquent pas beaucoup de variance.***

####**4. Mises en garde de l'APC**

+ ***Il y a quelques points à noter concernant l'APC. La première est qu'il peut être très difficile d'interpréter les composantes de l'ACP au-delà desquelles les composantes expliquent le plus la variance. L'ACP est plus une méthode de boîte noire que d'autres méthodes de réduction de la dimensionnalité. L'autre chose à noter est que l'ACP est une bonne étape à faire à la fin du parcours de prétraitement, en raison de la façon dont les données sont transformées et remodelées. Il peut être difficile de faire beaucoup de travail sur les fonctionnalités après l'ACP, autre que d'éliminer les composants qui ne sont pas utiles pour expliquer la variance.***
