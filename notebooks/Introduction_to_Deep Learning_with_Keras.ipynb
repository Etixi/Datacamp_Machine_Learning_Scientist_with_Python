####**2. Un réseau de neurones ?**

+ ***En un mot, un réseau de neurones est un algorithme d'apprentissage automatique qui est extrait avec des données d'apprentissage via sa couche d'entrée pour ensuite prédire une valeur à sa couche de sortie.***

####**3. Paramètres**

+ ***Chaque connexion d'un neurone à un autre a un poids associé, w. Chaque neurone, à l'exception de la couche d'entrée qui ne contient que la valeur d'entrée, a également un poids supplémentaire et nous l'appelons le poids de biais, b. Pendant l'anticipation, notre entrée est transformée par des multiplications et des additions de poids à chaque couche, la sortie de chaque neurone peut également être transformée par l'application de ce que nous appelons une fonction d'activation.***

####**4. Descente en dégradé**

+ ***L'apprentissage dans les réseaux de neurones consiste à ajuster les poids ou les paramètres pour donner la sortie souhaitée. Une façon d'y parvenir consiste à utiliser le célèbre algorithme de descente de gradient et à appliquer des mises à jour de poids de manière incrémentielle via un processus connu sous le nom de rétropropagation. C'était beaucoup de théorie ! Le code dans Keras est beaucoup plus simple comme nous allons le voir maintenant.***

####**3. Compilation**

+ ***Un modèle doit être compilé avant la formation. Nous pouvons compiler notre modèle en appelant la méthode compile dessus. La méthode de compilation reçoit un optimiseur, que nous pouvons voir comme l'algorithme qui sera utilisé pour mettre à jour nos poids de réseau de neurones, et une fonction de perte, qui est la fonction que nous voulons minimiser lors de la formation. Dans ce cas, nous choisissons ADAM comme optimiseur et l'erreur quadratique moyenne comme fonction de perte. Les optimiseurs et les fonctions de perte seront abordés plus tard dans le cours, alors ne vous en souciez pas pour le moment. La compilation de notre modèle ne produit aucune sortie. Notre modèle est maintenant prêt à s'entraîner !***

####**4. Formation**

+ ***Créer un modèle est inutile si nous ne le formons pas. Nous entraînons notre modèle en appelant la méthode d'ajustement et en transmettant les caractéristiques dans X_train, les étiquettes dans y_train et le nombre d'époques pour lesquelles s'entraîner. Au cours d'une époque, toutes nos données d'entraînement passent par le réseau et les mises à jour de poids respectives ont lieu par rétro-propagation. Au fur et à mesure que notre modèle est en cours de formation, nous obtiendrons une sortie montrant sa progression. Nous pouvons voir que le modèle s'améliore puisque la perte d'erreur quadratique moyenne diminue à chaque époque.***

####**5. Prédire**

+ ***Pour obtenir des prédictions à partir de notre modèle entraîné, il nous suffit d'appeler predict sur le nouvel ensemble de données. Nous pouvons stocker les prédictions dans une variable pour une utilisation ultérieure. Les prédictions ne sont que des nombres dans un tableau numpy, nous les interpréterons en fonction de notre ensemble de données et du problème à résoudre.***

####**6. Évaluer**

+ ***Pour évaluer rapidement les performances de notre modèle sur des données invisibles, nous pouvons utiliser la méthode d'évaluation du modèle. Cela effectue une anticipation avec tous les échantillons de notre jeu de données de test (X_test). `Le feed-forward consiste à calculer les sorties d'un modèle à partir d'un ensemble donné d'entrées. Il calcule ensuite l'erreur en comparant les résultats aux vraies valeurs stockées dans y_test`. Dans cet exemple particulier, le modèle que nous avons formé pendant 5 époques auparavant a une erreur quadratique moyenne de 0,25.***

####**2. Courbes d'apprentissage**

+ ***Jusqu'à présent, nous avons vu deux types de courbes d'apprentissage : les courbes de perte et les courbes de précision.***

####**3. Courbe de perte**

+ ***La perte a tendance à diminuer au fil des époques. Ceci est attendu puisque notre modèle apprend essentiellement à minimiser la fonction de perte. Les époques sont indiquées sur l'axe X et la perte sur l'axe Y. Au fil des époques, notre valeur de perte diminue. Après un certain nombre d'époques, la valeur converge, ce qui signifie qu'elle ne devient plus beaucoup plus basse que cela. Nous sommes arrivés au minimum.***

####**4. Courbe de précision**

+ ***Les courbes de précision sont similaires mais de tendance opposée si l'axe $Y$ indique la précision, elle a maintenant tendance à augmenter au fil des époques. Cela montre que notre modèle fait moins d'erreurs au fur et à mesure qu'il apprend.***

####**5. Sur-ajustement**

+ ***Si nous traçons les données de formation par rapport aux données de validation, nous pouvons identifier le surapprentissage. Nous verrons les courbes de formation et de validation commencer à diverger. Le surajustement se produit lorsque notre modèle commence à apprendre les particularités de nos données d'entraînement qui ne se généralisent pas bien sur des données invisibles. Le rappel d'arrêt précoce est utile pour arrêter notre modèle avant qu'il ne commence à sur-ajuster.***

####**6. Courbes instables**

+ ***Mais toutes les courbes ne sont pas lisses et jolies, nous trouverons souvent des courbes instables. De nombreuses raisons peuvent conduire à des courbes d'apprentissage instables ; l'optimiseur choisi, le taux d'apprentissage, la taille du lot, l'architecture du réseau, l'initialisation du poids, etc. Tous ces paramètres peuvent être ajustés pour améliorer nos courbes d'apprentissage du modèle, car nous visons une meilleure précision et une meilleure puissance de généralisation. ####**1. Taille des lots et normalisation des lots**

+ ***Il est temps d'apprendre les concepts de taille de lot et de normalisation de lot.***

####**2. Lots**

+ ***Un mini-lot est un sous-ensemble d'échantillons de données. Si nous entraînions un réseau de neurones avec des images, chaque image de notre ensemble d'apprentissage serait un échantillon et nous pourrions prendre des mini-lots de différentes tailles à partir du lot d'ensemble d'apprentissage.***

####**3. Mini-lot**

+ ***Rappelez-vous que pendant une époque, nous alimentons notre réseau, calculons les erreurs et mettons à jour les poids du réseau. Il n'est pas très pratique de mettre à jour nos pondérations de réseau une seule fois par époque après avoir examiné l'erreur produite par tous les échantillons d'apprentissage. En pratique, on prélève un mini-lot d'échantillons d'apprentissage. Et de cette façon, si notre ensemble d'entraînement a 9 images et que nous choisissons un batch_size de 3, nous effectuerons 3 mises à jour de poids par époque, une par mini-lot.***

####**4. Mini-lots**

+ ***Les réseaux ont tendance à s'entraîner plus rapidement avec des mini-lots car les poids sont souvent mis à jour. Parfois, les ensembles de données sont si volumineux qu'ils auraient du mal à tenir dans la mémoire RAM si nous n'utilisions pas de mini-lots. De plus, le bruit produit par une petite taille de lot peut aider à échapper aux minima locaux. Quelques inconvénients sont le besoin de plus d'itérations et de trouver une bonne taille de lot.***

####**5. Effets de la taille des lots**

+ ***Ici, vous pouvez voir comment différentes tailles de lots convergent vers un minimum au fil de la formation. L'entraînement avec tous les échantillons est affiché en bleu. Le mini-dosage est représenté en vert. La descente de gradient stochastique, en rouge, utilise un batch_size de 1. Nous pouvons voir comment le chemin vers la meilleure valeur pour nos poids est plus bruyant plus le batch_size est petit. Ils atteignent la même valeur après un nombre différent d'itérations.***

  + *pile d'échange*

####**6. Taille du lot à Keras**

+ ***Vous pouvez définir votre propre batch_size avec le paramètre batch_size sur la méthode d'ajustement du modèle. Keras utilise une taille de lot par défaut de 32. Des puissances croissantes de deux ont tendance à être utilisées. En règle générale, vous avez tendance à augmenter la taille de votre lot plus votre ensemble de données est grand.***

####**7. Normalisation dans l'apprentissage automatique**

+ ***La normalisation est une étape de prétraitement courante dans les algorithmes d'apprentissage automatique, en particulier lorsque les caractéristiques ont des échelles différentes. Une façon de normaliser les données consiste à soustraire sa valeur moyenne et à diviser par l'écart type. Nous avons toujours tendance à normaliser nos entrées de modèle. Cela évite les problèmes avec les fonctions d'activation et les gradients.***

####**8. Normalisation dans l'apprentissage automatique**

+ ***Cela laisse tout centré autour de 0 avec un écart type de 1.***

####**9. Raisons de la normalisation des lots**

+ ***La normalisation des entrées des réseaux de neurones améliore notre modèle. Mais les couches plus profondes sont formées en fonction des sorties de couche précédentes et puisque les poids sont mis à jour via la descente de gradient, les couches consécutives ne bénéficient plus de la normalisation et elles doivent s'adapter aux changements de poids des couches précédentes, trouvant plus de difficulté à apprendre leurs propres poids. La normalisation par lots garantit que, indépendamment des modifications, les entrées des couches suivantes sont normalisées. Il le fait de manière intelligente, avec des paramètres entraînables qui apprennent également dans quelle mesure cette normalisation est mise à l'échelle ou déplacée.***

####**10. Avantages de la normalisation par lots**

+ ***Cela améliore le flux de gradient, permet des taux d'apprentissage plus élevés, réduit la dépendance aux initialisations de poids, ajoute une régularisation à notre réseau et limite le changement de covariable interne ; qui est un drôle de nom pour la dépendance d'une couche vis-à-vis des sorties de la couche précédente lors de l'apprentissage de ses poids. La normalisation par lots est largement utilisée aujourd'hui dans de nombreux modèles d'apprentissage en profondeur.***

+ ***Construisez une fonction simple `create_model()` qui reçoit à la fois un taux d'apprentissage et une fonction d'activation comme arguments. L'optimiseur `Adam` a été importé en tant qu'objet de `tensorflow.keras.optimizers` afin que vous puissiez également modifier son paramètre de taux d'apprentissage.***

####**Instructions**

```python
# Creates a model given an activation and learning rate
def create_model(learning_rate=0.001, activation='relu'):
    # Create an Adam optimizer with the given learning rate
    opt = tf.keras.optimizers.legacy.Adam(learning_rate==learning_rate)

    # Create your binary classification model
    model = Sequential()
    model.add(Dense(128, input_shape=(30, ), activation=activation))
    model.add(Dense(256, activation=activation))
    model.add(Dense(1, activation='sigmoid'))

    # Compile your model with your optimizer, loss and metrics
    model.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
    return model
```

####**3. Démonstration des convolutions**

+ ***Elle consiste à appliquer un filtre, aussi appelé noyau, d'une taille donnée. Dans cette image, nous appliquons un noyau 3 par 3. Nous centrons la matrice de nombres du noyau lorsque nous glissons à travers chaque pixel de l'image, multipliant les valeurs du noyau et des pixels à chaque emplacement et faisant la moyenne de la somme des valeurs obtenues. Cela calcule effectivement une nouvelle image où certaines caractéristiques sont amplifiées en fonction du filtre utilisé. La sauce secrète des CNN consiste à laisser le réseau lui-même trouver les meilleures valeurs de filtre et à les combiner pour accomplir une tâche donnée.***

####**4. Architectures typiques**

####**2. Que sont les RNN ?**

####**3. Que sont les LSTM ?**

+ ***Les neurones LSTM sont assez complexes, ils sont en fait appelés unités ou cellules. Ils ont un état interne qui est transmis entre les unités, vous pouvez le voir comme une mémoire des étapes passées. Ensuite, il met à jour l'état et produit une nouvelle sortie qui est renvoyée, ainsi que transmise en tant qu'entrée à l'unité suivante.***

####**4. Que sont les LSTM ?**

+ ***Les unités LSTM effectuent plusieurs opérations. Ils apprennent ce qu'il faut ignorer, ce qu'il faut garder et sélectionner les informations passées les plus importantes afin de prédire l'avenir. Ils ont tendance à mieux fonctionner que les simples RNN pour la plupart des problèmes.***

```python
def predict_text(test_text, model = model):
  if len(test_text.split()) != 3:
    print('Text input should be 3 words!')
    return False

  # Turn the test_text into a sequence of numbers
  test_seq = tokenizer.texts_to_sequences([test_text])
  test_seq = np.array(test_seq)

  # Use the model passed as a parameter to predict the next word
  pred = model.predict(test_seq).argmax(axis = 1)[0]

  # Return the word that maps to the prediction
  return tokenizer.index_word[pred]
```
