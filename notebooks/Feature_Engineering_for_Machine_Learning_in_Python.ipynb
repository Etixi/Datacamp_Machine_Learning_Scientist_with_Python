####**1. Traiter les variables catégorielles**

+ ***Les variables catégorielles sont utilisées pour représenter des groupes de nature qualitative. Quelques exemples sont les couleurs, comme le bleu, le rouge, le noir, etc. ou le pays de naissance, comme l'Irlande, l'Angleterre ou les États-Unis. Bien que ceux-ci puissent être facilement compris par un humain, vous devrez encoder les caractéristiques catégorielles sous forme de valeurs numériques pour les utiliser dans vos modèles d'apprentissage automatique.***

####**2. Encodage des caractéristiques catégorielles**

+ ***A titre d'exemple, voici un tableau qui se compose du pays de résidence des différents répondants à l'enquête Stackoverflow. Pour passer des entrées qualitatives aux caractéristiques quantitatives, on peut naïvement penser qu'il suffirait d'attribuer un numéro à chaque catégorie dans une colonne, par exemple l'Inde pourrait être 1, les États-Unis 2, etc. Mais ces catégories ne sont pas ordonnées, donc l'attribution de cet ordre peut grandement pénaliser le l'efficacité de votre modèle. Ainsi, vous ne pouvez pas attribuer des nombres arbitraires à chaque catégorie car cela impliquerait une certaine forme d'ordre dans les catégories.***

####**3. Encodage des caractéristiques catégorielles**

+ ***Au lieu de cela, les valeurs peuvent être encodées en créant des caractéristiques binaires supplémentaires correspondant au fait que chaque valeur a été sélectionnée ou non, comme indiqué dans le tableau de droite. Ce faisant, votre modèle peut tirer parti des informations sur le pays donné, sans déduire aucun ordre entre les différentes options.***

####**4. Encodage des caractéristiques catégorielles**

+ ***Il existe deux approches principales pour représenter les colonnes catégorielles de cette manière, un encodage à chaud et un encodage factice. Ceux-ci sont très similaires et souvent confondus. En fait, par défaut, pandas effectue un encodage à chaud lorsque vous utilisez `la fonction get_dummies()`.***

####**5. Encodage à chaud**

+ ***L'encodage à chaud convertit $n$ catégories en $n$ fonctionnalités, comme illustré ici. Vous pouvez utiliser `la fonction get_dummies()` pour encoder des colonnes à chaud. La fonction prend un DataFrame et une liste de colonnes catégorielles que vous souhaitez convertir en une colonne encodée à chaud, et renvoie un DataFrame mis à jour avec ces colonnes incluses. Spécifier un préfixe avec l'argument préfixe peut améliorer la lisibilité comme la lettre C pour le pays a été utilisée ici.***

####**6. Encodage factice**

+ ***D'autre part, le codage factice crée $n-1$ caractéristiques pour $n$ catégories, en omettant la première catégorie. Notez que cette fois il n'y a pas de fonctionnalité pour la France, la première catégorie. Dans le codage factice, la valeur de base, la France dans ce cas, est codée par l'absence de tous les autres pays comme vous pouvez le voir sur la dernière ligne ici et sa valeur est représentée par l'ordonnée à l'origine. Pour l'encodage factice, vous pouvez utiliser la même fonction `get_dummies()` avec un argument supplémentaire, `drop_first` défini sur True comme indiqué ici.***

####**7. One-hot contre mannequins**

+ ***Ces deux méthodes présentent des avantages différents. L'encodage à chaud crée généralement des fonctionnalités beaucoup plus explicables, car chaque pays aura son propre poids qui peut être observé après la formation. Mais il faut être conscient qu'un encodage à chaud peut créer des caractéristiques entièrement colinéaires car les mêmes informations sont représentées plusieurs fois.***

####**8. One-hot contre mannequins**

+ ***Prenons par exemple une colonne catégorielle plus simple enregistrant le sexe des personnes interrogées. En enregistrant un 1 pour homme, l'information indiquant si la personne est une femme est déjà connue lorsque la colonne homme est 0. Cette double représentation peut entraîner une instabilité dans vos modèles et des valeurs fictives seraient plus appropriées.***

####**9. Limiter vos colonnes**

+ ***Cependant, l'encodage one-hot et l'encodage factice peuvent entraîner la création d'un grand nombre de colonnes s'il y a trop de catégories différentes dans une colonne. Dans ces cas, vous souhaiterez peut-être créer uniquement des colonnes pour les valeurs les plus courantes. Vous pouvez vérifier le nombre d'occurrences de différentes fonctionnalités dans une colonne à l'aide de la méthode `value_counts()` sur une colonne spécifique.***

####**10. Limiter vos colonnes**

+ ***Une fois que vous avez votre nombre d'occurrences, vous pouvez l'utiliser pour limiter les valeurs que vous allez inclure en créant d'abord un masque des valeurs qui se produisent moins de $n$ fois. Un masque est une liste de booléens décrivant les valeurs d'une colonne qui doivent être affectées. Nous trouvons d'abord les catégories qui apparaissent moins de $n$ fois à l'aide de l'attribut index et l'enveloppons dans `la méthode isin()`. Après avoir créé le masque, vous pouvez l'utiliser pour remplacer ces catégories qui apparaissent moins de $n$ fois par une valeur de votre choix, comme indiqué ici.***

####**2. Suppression par liste**

+ ***Si vous êtes convaincu que les valeurs manquantes dans votre ensemble de données se produisent au hasard (c'est-à-dire qu'elles ne sont pas intentionnellement omises), l'approche la plus efficace et la plus solide sur le plan statistique pour les traiter est appelée "analyse de cas complète" ou suppression par liste. Dans cette méthode, un enregistrement est entièrement exclu de votre modèle si l'une de ses valeurs est manquante. Prenons par exemple l'ensemble de données présenté ici. + ***Pour implémenter la suppression par liste à l'aide de pandas, vous pouvez utiliser la `méthode dropna()` , en définissant l'argument how sur `'any'. Cela supprimera toutes les lignes avec au moins une valeur manquante.***

####4. Suppression par liste en Python

####**5. Problèmes de suppression**

+ ***Bien que l'approche préférable dans les situations où les données manquantes se produisent de manière purement aléatoire est la suppression par liste, elle a ses inconvénients. ####**3. Mise à l'échelle Min-Max**

+ ***`La mise à l'échelle Min-Max` se produit lorsque vos données sont mises à l'échelle linéairement entre une valeur minimale et une valeur maximale, souvent 0 et 1, 0 correspondant à la valeur la plus basse de la colonne et 1 à la plus grande. ####**4. Mise à l'échelle Min-Max**

+ ***Alors qu'ici, après la mise à l'échelle min-max, bien que la distribution soit la même, les valeurs se situent entièrement entre 0 et 1.***

####**5. Mise à l'échelle Min-Max en Python**

####**6. Normalisation**

+ ***L'autre mesureur couramment utilisé est appelé `standardisation`. Au lieu de trouver une limite extérieure et de tout serrer à l'intérieur, la normalisation trouve plutôt la moyenne de vos données et centre votre distribution autour d'elle, en calculant le nombre d'écarts types par rapport à la moyenne de chaque point. Ces valeurs (le nombre d'écarts types) sont ensuite utilisées comme nouvelles valeurs. Cela centre les données autour de 0 mais n'a techniquement aucune limite aux valeurs maximales et minimales, comme vous pouvez le voir ici.***

####**2. Que sont les valeurs aberrantes ?**

+ ***Les valeurs aberrantes sont des points de données qui existent loin de la majorité de vos données. Cela peut se produire pour plusieurs raisons, telles qu'un enregistrement incorrect des données ou de véritables occurrences rares. Dans tous les cas, vous souhaiterez souvent supprimer ces valeurs car elles peuvent avoir un impact négatif sur vos modèles. ####**3. Détection basée sur les quantiles**

+ ***La première approche dont nous parlerons consiste à supprimer un certain pourcentage des valeurs les plus grandes et/ou les plus petites de vos données. Par exemple, vous pouvez supprimer les 5 % supérieurs. Ceci est réalisé en trouvant le 95ème quantile (le point en dessous duquel 95% de vos données résident) et en supprimant tout ce qui se trouve au-dessus. Cette approche est particulièrement utile si vous souhaitez éviter les valeurs les plus élevées de votre jeu de données. Lorsque vous utilisez cette approche, vous devez vous rappeler que même s'il n'y a pas de véritables valeurs aberrantes, vous supprimerez toujours les 5 % de valeurs les plus élevées de l'ensemble de données.***

####**5. Détection basée sur l'écart type**

Par exemple, vous souhaiterez peut-être éliminer toutes les données supérieures à 3 écarts-types par rapport à la moyenne car vous vous attendez à ce que ces points de données soient des valeurs aberrantes. Cette approche a l'avantage de ne supprimer que les valeurs véritablement extrêmes, par exemple si une seule valeur était une valeur aberrante, seule cette valeur serait affectée.***

####**6. Détection de l'écart type en Python**

+ ***Pour appliquer cela en Python, vous devez d'abord trouver la moyenne et l'écart type de votre colonne en appelant respectivement `les méthodes mean() et std()` sur la colonne. ####**Suppression des valeurs aberrantes statistiques**

+ ***Bien que la suppression des `N%` supérieurs de vos données soit utile pour s'assurer que les points très erronés sont supprimés, elle présente l'inconvénient de toujours supprimer la même proportion de points, même si les données sont correctes. Une approche alternative couramment utilisée consiste à supprimer les données situées à plus de trois écarts-types de la moyenne. Vous pouvez implémenter cela en calculant d'abord la moyenne et l'écart type de la colonne concernée pour trouver les limites supérieure et inférieure, et en appliquant ces limites en tant que masque au DataFrame. ####**Instructions**

+ Calculez l'écart type et la moyenne de la colonne ConvertedSalary de so_numeric_df.
+ Calculez les limites supérieure et inférieure à trois écarts-types de la moyenne dans les deux directions.

````python
# Find the mean and standard dev
std = so_numeric_df['ConvertedSalary'].std()
mean = so_numeric_df['ConvertedSalary'].mean()

# Calculate the cutoff
cut_off = std * 3
lower, upper = mean - cut_off, mean + cut_off

# Trim the outliers
trimmed_df = so_numeric_df[(so_numeric_df['ConvertedSalary'] < upper) & (so_numeric_df['ConvertedSalary'] > lower)]

# The trimmed box plot
trimmed_df[['ConvertedSalary']].boxplot()
plt.show()
````

####**1. Représentation Tf-Idf**

+ ***Bien que le nombre d'occurrences de mots puisse être un bon premier pas vers l'encodage de votre texte pour créer des modèles, il présente certaines limites. Le principal problème est que les nombres seront beaucoup plus élevés pour les éléments très courants, même lorsqu'ils se produisent dans tous les textes, ce qui leur confère peu de valeur en tant que caractéristique distinctive.***

####**2. Présentation de TF-IDF**

+ ***Prenons par exemple le décompte du mot "le" montré ici, avec de nombreuses occurrences dans chaque ligne. Pour empêcher ces mots courants de dominer votre modèle, une certaine forme de normalisation peut être utilisée. L'une des approches les plus efficaces pour ce faire est appelée `"Term Frequency Inverse Document Frequency" ou TF-IDF`.***

####**3. TF-IDF**

+ ***`TF-IDF divise le nombre de fois où un mot apparaît dans le document par une mesure de la proportion des documents où un mot apparaît dans tous les documents`. Cela a pour effet de réduire la valeur des mots courants, tout en augmentant le poids des mots qui n'apparaissent pas dans de nombreux documents.***

####**4. Importation du vectoriseur**

+ ***Pour utiliser un vectoriseur TF-IDF, l'approche est très similaire à la façon dont vous avez appliqué un vectoriseur de comptage. Vous devez d'abord importer `TfidfVectorizer() à partir de sklearn dot feature_extraction dot text`, puis vous l'attribuez à un nom de variable. Utilisons la télévision dans ce cas.***

####**5. Fonctionnalités maximales et mots vides**

+ ***Comme lorsque vous travailliez avec le vectoriseur Count où vous pouviez limiter le nombre d'entités créées en spécifiant des arguments lors de l'initialisation de TfidfVectorizer, vous pouvez spécifier le nombre maximum d'entités à l'aide de `max_features` qui n'utilisera que les 100 mots les plus courants. Nous spécifierons également le vectoriseur pour omettre un ensemble de `stop_words`, il s'agit d'une liste prédéfinie des mots anglais les plus courants tels que "and" ou "the". Bien que l'analyse des occurrences de mots individuels puisse être un moyen précieux de créer des caractéristiques à partir d'un morceau de texte, vous remarquerez que des mots individuels peuvent perdre tout leur contexte/signification lorsqu'ils sont visualisés indépendamment.***

####**Utilisation de n-grammes plus longs**
+ ***Jusqu'à présent, vous avez créé des fonctionnalités basées sur des mots individuels dans chacun des textes. Cela peut être assez puissant lorsqu'il est utilisé dans un modèle d'apprentissage automatique, mais vous craignez peut-être qu'en examinant les mots individuellement, une grande partie du contexte soit ignorée. Pour gérer cela lors de la création de modèles, vous pouvez utiliser des n-grammes qui sont des séquences de n mots regroupés. Par exemple:***

  + `bigrammes` : séquences de deux mots consécutifs
  + `trigrammes` : séquences de deux mots consécutifs

+ *Ceux-ci peuvent être créés automatiquement dans votre ensemble de données en spécifiant l'argument  ngram_range sous la forme d'un tuple $(n1, n2)$ où tous les n-grammes de la plage $n1$ à $n2$ sont inclus.*

####**Instructions**

+ Importer CountVectorizer depuis sklearn.feature_extraction.text.

````python
# Import CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer

# Instantiate a trigram vectorizer
cv_trigram_vec = CountVectorizer(max_features=100,
                                 stop_words='english',
                                  ngram_range=(3,3))

# Fit and apply trigram vectorizer
cv_trigram = cv_trigram_vec.fit_transform(speech_df['text_clean'])
````
